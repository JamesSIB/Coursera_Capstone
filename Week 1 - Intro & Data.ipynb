# The Battle of the Neighbourhoods - Intro/Business Problem

## 1. Introduction/Business Problem:

Pizza is one of the most popular foods - and rightly so. London is a popular destination for food lovers and in particular pizza lovers. There are an estimated 2000 pizza selling venues in London.

Through a recent survey it came to light that there were many pizza lovers in London that loved pizza and hated walking. These people weren't sure where to live to maximise this love and minimise this hate, and thus pizzalovers.com (completely fictious) was born. As stated pizzalovers.com caters to people who love pizza and hate walking. My typical customer is someone who wants to live in an area that has as many pizza places in as little walking time as possible.

The way we identify these areas is by a cluster analysis to show which areas have a high density of pizza places. We also give a quick ranking based on population figures of the borough - this is simply pizza places per capita.
